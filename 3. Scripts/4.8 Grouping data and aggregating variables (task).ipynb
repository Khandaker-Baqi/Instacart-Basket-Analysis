## Creating the loyalty flag

### Finding maximum "order_number" for each "user_id" with "transform()" function

```python
# finding maximum "order_number" for each "user_id" with "transform()" function
df_orders_products_merged['max_orders'] = df_orders_products_merged.groupby('user_id')['order_number'].transform(np.max)
```

### Checking the updated dataframe after executing the "transform()" function

```python
# checking the updated dataframe
df_orders_products_merged.head(50)
```

```python
# checking the frequecy distribution of the "max_orders" column
df_orders_products_merged['max_orders'].value_counts(dropna = False)
```

```python
# checking the "max_orders" column for null values
df_orders_products_merged['max_orders'].isnull().sum()
```

### Creating a loyalty flag based on the defined parameters

```python
# condition 1: order number greater than 40
df_orders_products_merged.loc[df_orders_products_merged['max_orders'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# condition 2: order number greater than 10 and less than or equal to 40
df_orders_products_merged.loc[(df_orders_products_merged['max_orders'] > 10) & (df_orders_products_merged['max_orders'] <= 40), 'loyalty_flag'] = 'Regular customer'
```

```python
# condition 3: order number less than or equal to 10
df_orders_products_merged.loc[df_orders_products_merged['max_orders'] <= 10, 'loyalty_flag'] = 'New customer'
```

### Checking the loyalty flag

```python
# checkign the 'loyalty_flag'
df_orders_products_merged['loyalty_flag'].value_counts(dropna = False)
```

## Descriptive analysis using agg() function while grouping by "loyalty_flag"

```python
# descriptive analysis using agg() function while grouping by "loyalty_flag"
df_orders_products_merged.groupby('loyalty_flag').agg({'prices' : ['mean', 'min', 'max', 'median']})
```

#### Comparison: The average prices of the products purchased by the loyal customers is lesser than the average prices of products purchased by other categories of customers. However, the minimum, maximum and median values are same for all categories

## Creating a spending flag

### Finding aggregate mean of "prices" while grouping by "user_id"

```python
# creating a new column with aggregate mean of "prices" while grouping by "user_id"
df_orders_products_merged['avg_price'] = df_orders_products_merged.groupby('user_id')['prices'].transform(np.mean)
```

### Checking the updated dataframe

```python
# checking the head of the updated dataframe
df_orders_products_merged.head(50)
```

```python
# checking the frequency distribution of the "avg_price" column
df_orders_products_merged['avg_price'].value_counts(dropna = False)
```

```python
# checking if there is any null valyue in the 'avg_price' column
df_orders_products_merged['avg_price'].isnull().sum()
```

### Creating a price flag based on the "avg_price" column

```python
# condition 1: average price of products purchased less than 10
df_orders_products_merged.loc[df_orders_products_merged['avg_price'] < 10, 'price_flag'] = 'Low spender'
```

```python
# condition 2: average price of products purchased higher than or equal to 10
df_orders_products_merged.loc[df_orders_products_merged['avg_price'] >= 10, 'price_flag'] = 'High spender'
```

### Checking the price flag

```python
# checking the updated dataframe
df_orders_products_merged.head(50)
```

```python
# checking the frequency distribution of the "price_flag" column
df_orders_products_merged['price_flag'].value_counts(dropna = False)
```

## Creating a frequency flag

### Finding the median of "days_since_prior_order" for every "user_id"

```python
# creating a new column with median of "days_since_prior_order" for each "user_id" using the "transform()" function
df_orders_products_merged['median_of_days_since_prior_order'] = df_orders_products_merged.groupby('user_id')['days_since_prior_order'].transform(np.median)
```

### Checking the updated dataframe

```python
# checking the head of the updated dataframe
df_orders_products_merged.head(50)
```

```python
# checking frequency distribution of the "median_of_days_since_prior_order" column
df_orders_products_merged['median_of_days_since_prior_order'].value_counts(dropna = False)
```

```python
# checking for any null vallue in the "median_of_days_since_prior_order" column
df_orders_products_merged['median_of_days_since_prior_order'].isnull().sum()
```

```python
# checking the records with "null" in median flag
df_orders_products_merged[df_orders_products_merged['median_of_days_since_prior_order'].isnull() == True]
```

### Creating a frequency flag based on the "median_of_days_since_prior_order" column

```python
# condition 1: median higher than 20
df_orders_products_merged.loc[df_orders_products_merged['median_of_days_since_prior_order'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
# condition 2: median higher than 10 and lower than or equal to 20
df_orders_products_merged.loc[(df_orders_products_merged['median_of_days_since_prior_order'] > 10) & (df_orders_products_merged['median_of_days_since_prior_order'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
# condition 3: median lower than or equal to 10
df_orders_products_merged.loc[df_orders_products_merged['median_of_days_since_prior_order'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# condition 4: median = null
df_orders_products_merged.loc[df_orders_products_merged['median_of_days_since_prior_order'].isnull() == True, 'frequency_flag'] = 'New customer'
```

### Checking the udated dataframe

```python
# checking the head of the updated dataframe
df_orders_products_merged.head(50)
```

```python
# checking frequency distribution of the "frequency_flag" column
df_orders_products_merged['frequency_flag'].value_counts(dropna = False)
```

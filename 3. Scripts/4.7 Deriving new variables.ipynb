### Subsettig the first 1 million rows as a new dataframe



```python
# subsetting the first 1 million rows as a new dataframe
df = df_ords_prods_merged[:1000000]
```

### Creating a user defined command for creating flags for different proce ranges to the first 1 million rows

```python
# creating user defined command
def price_label(row) :
    if row['prices'] <= 5 :
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15) :
        return 'Mid-range product'
    elif row['prices'] > 15 :
        return 'High-range product'
    else : 
        return 'Not enough data'
```

### Applying the user defuned function

```python
# applying the user defined function to the dataframe
df['price_range'] = df.apply(price_label, axis = 1)
```

### Checking the outcome of the user defined command

```python
# checking the price range flags
df['price_range'].value_counts(dropna = False)
```

## User defined functions with loc[] statement on the subset dataframe

```python
# setting flag to "High-range products" for products with price > 15
df.loc[(df['prices'] > 15), 'price_range_loc'] = 'High-range product'
```

```python
# setting flag to "Mid-range product" for products with price >=15 and <5
df.loc[(df['prices'] < 5) & (df['prices'] <= 15), 'price_range_loc'] = 'Mid-range product'
```

```python
# setting flag to "Low-range product" for products with price <= 5
df.loc[(df['prices'] <= 5), 'price_range_loc'] = 'Low-range product'
```
